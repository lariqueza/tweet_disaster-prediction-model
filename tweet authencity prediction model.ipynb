```python
#split the dataset
#from our originial dataframe we want to split into two so we can test it 
from sklearn.model_selection import train_test_split
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
#we split 33% into a test set, the larger portion will be used to 
df_train,df_test=train_test_split(df, test_size=0.33,random_state=1)
df_train.shape,df_test.shape
```

```python
df_test.head()
```

```python
df_train.head()
```

```python
#we need to drop the target column from the test set
test_data = df_test.drop(['target'], axis=1)
test_data.head()
```

```python
df_train
```

```python
#build a predictor
predictor= TabularPredictor(label ='target').fit(train_data = df_train, verbosity = 2,presets='best_quality')

#continue training the model with additional rounds using the same training data
#predictor.fit(train_data=df_train, verbosity=2, hyperparameters={'NN': {'num_epochs': 10}})
```

```python
predictor = TabularPredictor.load("AutogluonModels\\ag-20240123_152225\\")
```

```python
#shows us how many models autogluon trained to arrive at its conclusion
predictor.fit_summary()
```

```python
#find the best model selected by autogluon
predictor.leaderboard(df_train, silent=True)
```

```python
#check for any feature engineering enhancing
predictor.feature_importance(data=df_train)
```

```python
#we feed the test data to the classifier for prediction and we can store it in a DataFrame
y_pred = predictor.predict(test_data)
y_pred=pd.DataFrame(y_pred,columns=['target'])
y_pred
```

```python
#To understand the evaluation metric ‘accuracy’, let us print the details for it
predictor.evaluate(df_test)
```
